```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Examples of MVN

```python
def plot_MVN (m): 
  plt.scatter(m[:, 0], m[:, 1], alpha=0.2, s=10)
  plt.xlabel('X1')
  plt.ylabel('X2')
  plt.grid(True)
  plt.axis('equal')
  # plt.show()
  return plt
```

```python
def plot_each_coord (m) :
  fig, axs = plt.subplots(1, 2, figsize=(12, 6))

  sns.kdeplot(np.array(m)[:, 0], fill=True, ax=axs[0])
  axs[0].set_title('X1')

  sns.kdeplot(np.array(m)[:, 1], fill=True, ax=axs[1])
  axs[0].set_title('X2')

  plt.tight_layout()
  plt.show()
```

```python

Mean1 = [0, 0]
Sigma1 = [[1, 0], [0, 1]]

m1 = np.random.multivariate_normal(mean=Mean1, cov=Sigma1, size=5000)

plot_MVN(m1).show()
```

```python
plot_each_coord (m1) 
```

```python
Mean2 = [0, 0]
Sigma2 = [[4, 0], [0, 1]]

m2 = np.random.multivariate_normal(mean=Mean2, cov=Sigma2, size=5000)

plot_MVN(m2).show()
```

```python
plot_each_coord(m2)
```

### Relation to Eigenvalues and Eigenvectors

```python
Mean3 = [0, 0]
Sigma3 = [[2, -1], [-1, 2]]

m3 = np.random.multivariate_normal(mean=Mean3, cov=Sigma3, size=5000)

plt3 = plot_MVN(m3)
plt3.show()
```

```python
# eigenvalues and eigenvectors of Sigma3
_, eigenvectors = np.linalg.eig(Sigma3)
plt3 = plot_MVN(m3)

# plot the eigenvectors
plt3.quiver(0, 0, eigenvectors[0][0], eigenvectors[0][1], angles='xy', scale_units='xy', scale=0.5, color='red')
plt3.quiver(0, 0, eigenvectors[1][0], eigenvectors[1][1], angles='xy', scale_units='xy', scale=0.5, color='red')
```

```python
Mean4 = [0, 0]
Sigma4 = [[1, 1], [1, 1]]

m4 = np.random.multivariate_normal(mean=Mean4, cov=Sigma4, size=1000)

plot_MVN(m4)
```

```python
Mean5 = [0, 0]
Sigma5 = [[2, -1], [-1, 1]]

m5 = np.random.multivariate_normal(mean=Mean5, cov=Sigma5, size=1000)

plot_MVN(m5).show()
```

## Normal in each coordinate but not MVN

```python
def pieceWise (x):
  if abs(x) <= 1:
    return -x
  else:
    return x
  

n6 = np.random.normal(0, 1, 1000)

m6 = np.array([[x, pieceWise(x)] for x in n6])
```

```python
plot_each_coord(m6)
```

```python
plot_MVN(m6).show()
```
